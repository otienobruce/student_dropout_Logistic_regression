We go on to map the columns

```python
df['gender'] = df.gender.map({'Female':0,'Male':1})
df['Partner'] = df.Partner.map({'No':0,'Yes':1})
df['Dependents'] = df.Dependents.map({'No':0,'Yes':1})
df['PhoneService'] = df.PhoneService.map({'No':0,'Yes':1})
df.PaperlessBilling = df.PaperlessBilling.map({'No':0,'Yes':1})
df.Churn = df.Churn.map({'No':0,'Yes':1})
```

```python
df = pd.get_dummies(df,columns=['PaymentMethod','Contract','StreamingMovies','StreamingTV','TechSupport','DeviceProtection','OnlineBackup','OnlineSecurity','InternetService','MultipleLines'],drop_first=True)
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['Churn'])
y = df.Churn
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,stratify=y,test_size=0.2)
```

```python
(X_train == ' ').sum()
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression(
    max_iter=2000,
    random_state=42,
    class_weight='balanced'
)
```

```python
model.fit(X_train_scaled,y_train)
```

```python
model.predict(X_test_scaled)
```

```python
model.score(X_test_scaled,y_test)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
coef_df = pd.DataFrame({
    'Features':X.columns,
    'Coefficient': model.coef_[0]
})
coef_df
```

```python
scaler = StandardScaler()
```

```python
x = df.drop(columns=['PhoneService','PaymentMethod_Credit card (automatic)','Partner','Churn'])
y = df.Churn
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,stratify=y,test_size=0.2)
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

model_l1 = LogisticRegression(
    penalty='l1',
    solver='liblinear',
    class_weight='balanced',
    max_iter=2000
)

model_l1.fit(x_train_scaled, y_train)
```

```python
model_l1.predict(x_test_scaled)
```

```python
model_l1.score(x_test_scaled,y_test)
```

```python
y_pred = model_l1.predict(x_test_scaled)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
import pandas as pd

coef = pd.DataFrame({
    "Feature": x.columns,
    "Coefficient": model_l1.coef_[0]
})

coef
```

## Model Interpretation

Key drivers of churn include:

• Tenure (longer tenure reduces churn)
• InternetService_Fiber optic
• MonthlyCharges
• Contract length

Customers with short tenure and high monthly charges are more likely to churn.
